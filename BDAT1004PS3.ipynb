#Question 1


## Step 1: Import the necessary libraries

```python
import pandas as pd
```

## Step 2: Import the dataset from the provided address

```python
dataset_address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

## Step 3. Assign it to a variable called users

```python
users = pd.read_csv(dataset_address, sep='|')
users.head()
```

## Step 4: Discover what is the mean age per occupation

```python
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print(mean_age_per_occupation)
```

## Step 5: Discover the Male ratio per occupation and sort it from the most to the least

```python
male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)
print(male_ratio_per_occupation)
```

## Step 6: For each occupation, calculate the minimum and maximum ages

```python
min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max_ages_per_occupation)
```

## Step 7: For each combination of occupation and sex, calculate the mean age

```python
mean_age_per_occupation_and_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_per_occupation_and_gender)
```

## Step 8: For each occupation, present the percentage of women and men

```python
total_per_occupation = users.groupby('occupation').size()
percentage_per_occupation = users.groupby(['occupation', 'gender']).size() / total_per_occupation * 100
print(percentage_per_occupation)
```

```python
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
data.head()
```

## Step 3: Create the line plot

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Marriages_per_1000'], label='Marriages per 1000')
plt.plot(data['Year'], data['Divorces_per_1000'], label='Divorces per 1000')

# Label the axes
plt.xlabel('Year')
plt.ylabel('Number per 1000 people')

# Add a title
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

## Step 4:  filter for the years 1900, 1950, and 2000

```python
years_of_interest = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years_of_interest)]
print(filtered_data)
```

##  Step 5:  Calculate the per capita values

```python
filtered_data['Marriages per 1000'] = filtered_data['Marriages'] / filtered_data['Population'] * 1000
filtered_data['Divorces per 1000'] = filtered_data['Divorces'] / filtered_data['Population'] * 1000
```

## Step 6: Create the vertical bar chart

```python

plt.figure(figsize=(8, 6))
bar_width = 0.35
index = range(len(years_of_interest))

marriages_bars = plt.bar(index, filtered_data['Marriages per 1000'], bar_width, label='Marriages per 1000')
divorces_bars = plt.bar([i + bar_width for i in index], filtered_data['Divorces per 1000'], bar_width, label='Divorces per 1000')

# Label the axes and add a title
plt.xlabel('Year')
plt.ylabel('Number per 1000 people')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')

# Add tick labels for the x-axis
plt.xticks([i + bar_width / 2 for i in index], years_of_interest)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
